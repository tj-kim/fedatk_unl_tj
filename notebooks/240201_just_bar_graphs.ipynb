```python
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

plt.rc('font', **font)
plt.figure(dpi=80, figsize=(6,4))

# Sample data for three groups (replace with your actual data)
group_labels = ['ARU', 'Label Flip', 'Noise Injection']
data_a = [0.8337,	0.59,	0.0956]  # Test Acc FAT
data_b = [0.01,	0.1284,	0.0956]  # ADV Acc FAT
data_c = [0.84406,	0.78875,	0.5409]  # Data for bars labeled C
data_d = [0.0684,	0.221,	0.182]  # Data for bars labeled D

# Calculate standard deviations for each group (replace with your actual data)
std_a = [0.0585,	0.0983,	0.069]
std_b = [0.0128,	0.05,	0.069]
std_c = [0.052,	0.077,	0.063]
std_d = [0.0285,	0.0537,	0.0568]

# Bar width
bar_width = 0.2

# Set up positions for the bars
index = np.arange(len(group_labels))
bar_a_positions = index - 1.5 * bar_width
bar_b_positions = index - 0.5 * bar_width
bar_c_positions = index + 0.5 * bar_width
bar_d_positions = index + 1.5 * bar_width

# Plotting bars with standard deviation error bars and hatch patterns
plt.bar(bar_a_positions, data_a, width=bar_width, label='Test Acc. (FAT)', yerr=std_a, capsize=5, color='blue')
plt.bar(bar_b_positions, data_b, width=bar_width, label='Adv. Acc (FAT)', yerr=std_b, capsize=5, color='orange')
plt.bar(bar_c_positions, data_c, width=bar_width, label='Test Acc. (TM)', yerr=std_c, capsize=5, color='blue', hatch='//')
plt.bar(bar_d_positions, data_d, width=bar_width, label='Adv. Acc (TM)', yerr=std_d, capsize=5, color='orange', hatch='//')

# Set up labels and title
plt.ylabel('Accuracy')
plt.title('ARU Comparison to Other Poisoning Attacks')

# Set up group labels
plt.xticks(index, group_labels)

# Add legend
plt.legend(fontsize='x-small')
plt.ylim(bottom=0)

# Show the plot
plt.show()
```
