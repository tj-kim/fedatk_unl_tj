```python
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

plt.rc('font', **font)
plt.figure(dpi=80, figsize=(6,4))

# Sample data for four groups (replace with your actual data)
group_labels = ['Trim Mean \n i.i.d.', 'Trim Mean \n non-i.i.d.', 'Median \n i.i.d.', 'Median \n non-i.i.d.']
data_a = [0.78, 0.81, 0.8, 0.81]  # Data for bars labeled A
data_b = [0.42, 0.27, 0.53, 0.34]  # Data for bars labeled B

# Calculate standard deviations for each group (replace with your actual data)
std_a = [0.02, 0.02, 0.03, 0.02]
std_b = [0.06, 0.02, 0.05, 0.04]

# Bar width
bar_width = 0.35

# Set up positions for the bars
index = np.arange(len(group_labels))
bar_a_positions = index
bar_b_positions = index + bar_width

# Plotting bars with standard deviation error bars
plt.bar(bar_a_positions, data_a, width=bar_width, label='Test Acc.', yerr=std_a, capsize=5)
plt.bar(bar_b_positions, data_b, width=bar_width, label='Adv. Acc.', yerr=std_b, capsize=5)

# Set up labels and title
# plt.xlabel('Groups')
plt.ylabel('Accuracy')
plt.title('Data Distribution and Robust Aggregation')

# Set up group labels
plt.xticks(index + bar_width / 2, group_labels)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
